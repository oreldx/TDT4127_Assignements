# Theory - Numerics exercise 3

### a)
How can one deduce the secant method from Newton's method

We are starting from the Newton's method equation : xn+1 = xn - f(xn)/f'(xn)
By replacing the derivative with the Taylor approximation of the derivative we are getting by isolating it in : f(xn) = f(xn+1) + (xn - xn+1) f'(xn+1)
We are obtaining the same equation as Newton's method, without the derivative.

### b)
Will Newton's method always converge to a zero?

There is different cases where the Newton's method could not converge to a zero, some of them are:
 - if there is stationary point, or other words, if the derivative is equal to 0, we aren't able to use the equation (because of the 0 division)
 - if the derivative is too close from 0, it could lead to an excessive perpetual overshoot and then never converge

### c)
If no derivative is available, which method should you choose - Newton's method or the secant method?

The secant method because we are using the Taylor approximation which doesn't need the actual derivative.
